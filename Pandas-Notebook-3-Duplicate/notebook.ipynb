### 129. Write a function that takes the employee_salary and returns:

1. "High" if the salary is >= 600000
2. "Medium" if the salary is > 300000 and < 600000
3. "Low" otherwise.

Hint: Use the apply() method.

**Store the returned series to a new column 'Salary_category'.**

```python
salary_category=[]
def salary_categories(x):
    for i in x:
        if i>=600000:
            salary_category.append('high')
        elif (i>300000 & i<600000):
            salary_category.append('Medium')
        else:
            salary_category.append('Low')
    return salary_category

salary_category=salary_categories(df['Employee_Salary'])
df['Salary_Category'] = salary_category
```
